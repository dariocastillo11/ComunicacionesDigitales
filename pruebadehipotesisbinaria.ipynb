```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Configuración de parámetros
n = 50  # Número de muestras
p_null = 0.5  # Probabilidad según la hipótesis nula
p_alt = 0.7  # Probabilidad según la hipótesis alternativa
observed_successes = 30  # Éxitos observados

# Distribuciones binomiales
x = np.arange(0, n + 1)
null_dist = binom.pmf(x, n, p_null)
alt_dist = binom.pmf(x, n, p_alt)

# Gráficos
plt.figure(figsize=(12, 6))

# Gráfico de la hipótesis nula
plt.subplot(1, 2, 1)
plt.bar(x, null_dist, color='blue', alpha=0.6, label='Hipótesis nula (p=0.5)')
plt.axvline(observed_successes, color='red', linestyle='--', label='Éxitos observados (35)')
plt.title('Distribución según Hipótesis Nula')
plt.xlabel('Número de éxitos')
plt.ylabel('Probabilidad')
plt.legend()

# Gráfico de la hipótesis alternativa

plt.bar(x, alt_dist, color='green', alpha=0.6, label='Hipótesis alternativa (p=0.7)')
plt.axvline(observed_successes, color='red', linestyle='--', label='Éxitos observados (35)')
plt.title('Distribución según Hipótesis Alternativa')
plt.xlabel('Número de éxitos')
plt.ylabel('Probabilidad')
plt.legend()

# Mostrar gráficos
plt.tight_layout()
plt.show()

# Prueba de hipótesis
p_value = sum(null_dist[observed_successes:])
print(f"Valor p para la hipótesis nula: {p_value:.4f}")
```
